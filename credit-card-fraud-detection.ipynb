# Model Creation and comparision with other models

```python
x = df.drop(["Class","Amount"], axis = 1)
y = df["Class"]
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import f1_score, accuracy_score,precision_score, recall_score
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2,random_state = 42)
```

```python
acc = []
cv_scores = []
f1_scores=[]
prec=[]
rec=[]
def model(model):
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    Acc = accuracy_score(y_test, y_pred)
    print("Accuracy: {}".format(Acc))
    acc.append(Acc)
    cv_score = cross_val_score(model,x_train,y_train,cv=5)
    print("Cross Val Score: {}".format(np.mean(cv_score)))
    cv_scores.append(np.mean(cv_score))
    Prec = precision_score(y_test, y_pred)
    print("The precision is {}".format(Prec))
    prec.append(Prec)
    Rec = recall_score(y_test, y_pred)
    print("The recall is {}".format(Rec))
    rec.append(Rec)
    F1_score = f1_score(y_test, y_pred)
    print("The F1-Score is {}".format(F1_score))
    f1_scores.append(F1_score)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
model(clf)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
model(clf)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
model(clf)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
model(clf)
```

```python
from sklearn.svm import SVC
clf = SVC()
model(clf)
```

```python
models = ["LogisticRegression","RandomForestClassifier","DecisionTreeClassifier","KNeighborsClassifier","SVC"]
scores = { "Model Name" : models , "Cross Validation Score" : cv_scores, "Precision" : prec , "Recall" : rec, "F1 Score": f1_scores}
df1 = pd.DataFrame(scores)
df1
```

# Handling Imbalanced Data via Near Miss Algorithm

```python
from imblearn.under_sampling import NearMiss
nm = NearMiss()
x_nm, y_nm = nm.fit_resample(x, y)
print(x_nm.shape,y_nm.shape)
```

# Standarization

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
x_scaled = scalar.fit_transform(x_nm)
```

# Model Creation and comparision with other models

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import f1_score, accuracy_score,precision_score, recall_score
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y_nm, test_size = 0.2,random_state = 42)
```

```python
acc1 = []
cv_scores1 = []
f1_scores1=[]
prec1=[]
rec1=[]
def model1(model):
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    Acc = accuracy_score(y_test, y_pred)
    print("Accuracy: {}".format(Acc))
    acc1.append(Acc)
    cv_score1 = cross_val_score(model,x_train,y_train,cv=5)
    print("Cross Val Score: {}".format(np.mean(cv_score1)))
    cv_scores1.append(np.mean(cv_score1))
    Prec = precision_score(y_test, y_pred)
    print("The precision is {}".format(Prec))
    prec1.append(Prec)
    Rec = recall_score(y_test, y_pred)
    print("The recall is {}".format(Rec))
    rec1.append(Rec)
    F1_score = f1_score(y_test, y_pred)
    print("The F1-Score is {}".format(F1_score))
    f1_scores1.append(F1_score)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
model1(clf)
```
